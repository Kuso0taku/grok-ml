```python
# import libs
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# read data
data = np.genfromtxt("test.csv", delimiter=',', skip_header=1)
rooms = [int(i[1]) for i in data]
prices = [i[-1] for i in data]
```

```python
# data to be plotted
x = rooms
y = prices

plt.scatter(x, y, s=50, marker='o')

# plotting
plt.xlabel("Rooms")
plt.ylabel("Prices")
plt.xticks(range(0, 10))
plt.yticks(range(0, 600, 100))
plt.show()
```
