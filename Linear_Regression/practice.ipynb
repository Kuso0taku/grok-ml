```python
# import libs
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# read data
data = np.genfromtxt("test.csv", delimiter=',', skip_header=1)
rooms = [int(i[1]) for i in data]
prices = [i[-1] for i in data]
```

```python
# data to be plotted
x = rooms
y = prices

plt.scatter(x, y, s=50, marker='o')

# plotting
plt.xlabel("Rooms")
plt.ylabel("Prices")
plt.xticks(range(0, 10))
plt.yticks(range(0, 600, 100))
plt.show()
```

```python
from linreg import Linear_regression

model_simple = Linear_regression(rate=0.006, method=0) # приходится подбирать
model_quad = Linear_regression(rate=0.01, method=1)
model_absolute = Linear_regression(method=2)

model_simple.fit(rooms, prices) # требует тонкой настройки и абсолютно неэффективен
model_quad.fit(rooms, prices) # сходится быстрее всех (заметно на небольшом количестве эпох)
model_absolute.fit(rooms, prices) # немного медленнее
```

```python
print(model_simple.predict(4))
print(model_quad.predict(4))
print(model_absolute.predict(4))
```

```python
plt.scatter(x, y, s=50, marker='o', color="grey") # dots

# plotting
plt.xlabel("Rooms")
plt.ylabel("Prices")
plt.xticks(range(0, 10))
plt.yticks(range(0, 600, 100))
plt.plot(rooms, [model_simple.predict(i) for i in rooms], color="red")
plt.plot(rooms, [model_quad.predict(i) for i in rooms], color="blue")
plt.plot(rooms, [model_absolute.predict(i) for i in rooms], color="green")
plt.show()
```
